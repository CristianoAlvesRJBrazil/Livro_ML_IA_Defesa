## 10.1 Marcos internacionais emergentes

**Não existe, hoje, um regime global único e vinculante para a IA militar.** O que há é um **mosaico** de instrumentos — de dureza jurídica variável, por vezes sobrepostos, sempre em evolução. Três pilares o estruturam:

- A **Convenção sobre Certas Armas Convencionais (CCW)** da ONU é o foro internacional central para os sistemas de armas letais autônomos (*LAWS*). Seu **Grupo de Peritos Governamentais (GGE)** adotou, em 2019, princípios orientadores — entre eles, que **o DIH se aplica integralmente** a tais sistemas e que **a responsabilidade humana pelo emprego da força deve ser preservada**. Por operar por consenso, avança devagar e reflete divergências profundas;
- A **OTAN** adotou princípios de uso responsável: licitude, responsabilidade e prestação de contas, explicabilidade e rastreabilidade, confiabilidade, governabilidade e mitigação de vieses. São compromissos de ***soft law*** — orientam, sem obrigar como um tratado;
- O **AI Act europeu** é a primeira regulação horizontal abrangente de IA, baseada em risco. Aqui, uma ressalva crucial: **ele exclui de seu escopo os sistemas destinados exclusivamente a fins militares, de defesa ou de segurança nacional**. A IA militar permanece governada por outros instrumentos — o DIH e o direito nacional.

> **📝 Nota** — A fragmentação tem uma consequência prática: o profissional de defesa opera sob **múltiplos marcos**, de forças distintas e em evolução — e, acima de todos eles, sob o **direito internacional humanitário**, que vincula *independentemente* de qualquer tratado específico sobre IA. Conhecer o mosaico é necessário; mas o **solo firme**, sobre o qual não há dúvida de aplicabilidade, é o DIH (Seção 10.3).

## 10.2 O princípio do controle humano significativo

No centro de todo o debate está uma ideia: a de que **os seres humanos devem manter controle significativo sobre o uso da força** — que decisões de vida e morte não podem ser plenamente delegadas a uma máquina. É o **controle humano significativo** (*meaningful human control*), e é o fio que este livro perseguiu do primeiro capítulo ao último.

O que torna o controle *significativo*, e não meramente *nominal*? **Não basta** um humano formalmente "no circuito" que apenas homologa a sugestão da máquina — a figura que o **viés de automação** (Capítulos 4 e 5) esvazia. Exige-se que o operador disponha de:

- **compreensão suficiente** da situação (não uma caixa-preta);
- **tempo para deliberar**, compatível com a decisão;
- **capacidade de intervir** — de vetar, corrigir ou abortar — a qualquer momento;
- e que a decisão seja **registrada** e passível de responsabilização.

O conceito é central, mas **não há acordo preciso** sobre a sua definição. A própria expressão é disputada: muitos adotam *"controle humano significativo"*; os EUA preferem *"níveis apropriados de julgamento humano"*. Discute-se também **onde** no ciclo o controle deve residir — o humano *no* circuito (**in the loop**), *sobre* o circuito (**on the loop**) ou *fora* dele (**out of the loop**).

> **🛡️ Contexto de defesa** — O controle humano significativo tem **exigências técnicas**, não apenas doutrinárias. Para que o humano decida de fato, o sistema precisa **apresentar informação compreensível**, **conceder tempo de deliberação**, **permitir intervenção a qualquer momento** e **registrar a decisão**. Cada uma recai sobre o *projeto* do sistema — e é o que a governança técnica da Seção 10.5 materializa. **Um princípio ético sem mecanismo que o realize é uma intenção, não uma salvaguarda.**

```python
# Os tres graus de controle humano no ciclo de decisao
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

niveis = [
    ("IN the loop", "o humano AUTORIZA\ncada acao", "tab:green",
     "controle maximo:\na maquina espera o humano"),
    ("ON the loop", "o humano SUPERVISIONA\ne pode intervir/vetar", "tab:orange",
     "a maquina age;\no humano aborta se preciso"),
    ("OUT of the loop", "a maquina age\nSEM intervencao humana", "tab:red",
     "sem controle significativo:\no ponto que o livro rejeita"),
]
fig, ax = plt.subplots(figsize=(11, 3.4))
for i, (titulo, papel, cor, nota) in enumerate(niveis):
    x = i * 3.6
    ax.add_patch(FancyBboxPatch((x, 1.2), 3.0, 1.5, boxstyle="round,pad=0.1",
                                facecolor=cor, alpha=0.25, edgecolor=cor, lw=2))
    ax.text(x + 1.5, 2.35, titulo, ha="center", weight="bold", fontsize=12)
    ax.text(x + 1.5, 1.75, papel, ha="center", fontsize=9.5)
    ax.text(x + 1.5, 0.7, nota, ha="center", fontsize=8.5, style="italic",
            color="0.3")
    if i < 2:
        ax.add_patch(FancyArrowPatch((x + 3.05, 1.95), (x + 3.55, 1.95),
                                     arrowstyle="->", mutation_scale=20, lw=1.5))
ax.text(5.4, 3.15, "menos autonomia da maquina  <-------->  mais autonomia",
        ha="center", fontsize=9, color="0.4")
ax.set_xlim(-0.3, 11); ax.set_ylim(0.2, 3.4); ax.axis("off")
plt.title("Onde reside o controle humano? A tese do livro: IN / ON, nunca OUT",
          fontsize=12)
plt.tight_layout(); plt.show()

print("A tese da obra: a maquina percebe, tria e recomenda;")
print("a decisao sobre o uso da forca permanece HUMANA, informada e responsavel.")
```

**Observe:** a distinção não é acadêmica. Um humano *out of the loop* — ou um *in the loop* que apenas carimba — não exerce controle significativo. O projeto do sistema é o que determina em qual coluna ele efetivamente cai: sem informação compreensível, tempo e capacidade de veto, o "controle" é nominal.

## 10.3 Direito internacional humanitário aplicado a sistemas autônomos

O **direito internacional humanitário** (DIH), ou direito dos conflitos armados, vincula toda arma e todo emprego da força, autônomos ou não. Seus princípios cardeais, aplicados a sistemas autônomos, revelam **tanto a régua quanto a dificuldade**:

- A **distinção** exige separar combatentes de civis, e objetivos militares de bens civis. É o teste mais duro para uma máquina — e **aqui os limites técnicos dos Capítulos 4 e 8 tornam-se limites jurídicos**: um detector iludido por um adesivo adversarial, ou frágil a um deslocamento de domínio, não é apenas um problema de engenharia — é uma **potencial violação da distinção**;
- A **proporcionalidade** proíbe que o dano civil esperado seja excessivo frente à vantagem militar concreta — um juízo contextual, difícil de codificar;
- A **precaução** obriga a tomar todas as medidas factíveis para poupar civis;
- E a **cláusula de Martens** invoca os "princípios da humanidade" e os "ditames da consciência pública" — **nem tudo o que é tecnicamente possível é juridicamente ou moralmente admissível**.

Sobre todos paira a **lacuna de responsabilização** (*accountability gap*). Se um sistema autônomo comete uma violação, **quem responde** — o comandante que o empregou, o operador que o supervisionava, quem o programou, quem o fabricou? O DIH exige que haja responsabilidade; **a autonomia tende a difundi-la**. E é aqui que a disciplina técnica do livro reencontra a ética: *não há responsabilização sem rastreabilidade, e não há rastreabilidade sem a verificação, a validação e o registro* (Capítulo 8 e Seção 10.5).

> **⚠️ Armadilha comum** — **Supor que a conformidade com o DIH é um problema resolvido no plano das regras.** Ela depende, de forma inescapável, das **propriedades técnicas** do sistema. Um modelo que não distingue com robustez, que não é auditável ou cujo comportamento não é rastreável, **viola o DIH na prática**, por mais impecável que seja a doutrina que o emprega. A robustez adversarial (Capítulo 8), o envelope operacional e a rastreabilidade não são refinamentos de engenharia: são **pré-condições de licitude**. *Fragilidade técnica é, aqui, risco jurídico e moral.*

A célula abaixo mapeia cada princípio do DIH à **propriedade técnica** — construída ao longo do livro — de que a sua satisfação depende (o Exercício 10.4 desenvolve esse mapa):

```python
# Cada principio do DIH depende de uma propriedade tecnica do livro
mapa_dih = [
    ("Distincao", "separar combatentes de civis, alvos de bens civis",
     "deteccao robusta (Cap. 4) + robustez adversarial (Cap. 8): um "
     "detector iludido viola a distincao"),
    ("Proporcionalidade", "dano civil nao excessivo vs. vantagem militar",
     "avaliacao da CAUDA, nao da media (Caps. 7-8): o pior caso e' que "
     "determina o dano possivel"),
    ("Precaucao", "todas as medidas factiveis para poupar civis",
     "envelope operacional + degradacao graciosa (Caps. 8-9): nao empregar "
     "fora das condicoes validadas"),
    ("Responsabilidade", "haver sempre alguem que responde",
     "trilha de auditoria + ficha de modelo (Sec. 10.5): sem rastro, nao ha "
     "responsabilizacao"),
]
for principio, exigencia, tecnica in mapa_dih:
    print(f"[ {principio.upper()} ]  {exigencia}")
    print(f"   -> depende de: {tecnica}\n")
print("Limites TECNICOS viram limites JURIDICOS: fragilidade e' risco de licitude.")
```

**Observe:** a coluna da direita é este livro inteiro. A distinção depende da detecção robusta do Capítulo 4 e da resistência adversarial do Capítulo 8; a proporcionalidade, da avaliação da cauda dos Capítulos 7–8; a precaução, do envelope operacional e da degradação graciosa dos Capítulos 8–9; a responsabilidade, da governança que veremos a seguir. **A ética não paira sobre a engenharia — repousa sobre ela.**

## 10.5 Governança técnica: auditabilidade, rastreabilidade e *accountability*

**Princípios são vazios sem os mecanismos que os realizam.** A governança da IA de defesa não se decreta apenas em doutrina: **constrói-se dentro do sistema**. Três propriedades técnicas, entrelaçadas, a sustentam:

- A **auditabilidade** é a capacidade de examinar e compreender as decisões *após o fato* — o que exige registro, explicabilidade e acesso;
- A **rastreabilidade** é a capacidade de **reconstituir uma decisão** através do modelo que a produziu, de suas entradas, de sua versão, dos dados que o treinaram e do humano que interveio;
- A ***accountability*** (responsabilização) é a atribuição clara de responsabilidade, **só possível sobre as duas anteriores**: *não se responsabiliza alguém por uma decisão que não se pode rastrear nem compreender.*

Essas propriedades **costuram todo o livro**: a reprodutibilidade (sementes, versionamento — Capítulos 3 e 7); a verificação, a validação e o envelope operacional (Capítulo 8); a documentação e a trilha de decisão. A **Listagem 10.1** implementa uma **trilha de auditoria** de decisões:

```python
# Listagem 10.1 - Trilha de auditoria: registro rastreavel de cada decisao.
def registrar_decisao(caminho, modelo_versao, entrada, saida,
                      confianca, decisao_humana):
    # Trilha de auditoria: registra cada decisao apoiada por modelo,
    # com o suficiente para RASTREAR e RESPONSABILIZAR depois.
    evento = {
        "instante": time.time(),
        "modelo_versao": modelo_versao,          # qual modelo decidiu
        "entrada_hash": hashlib.sha256(          # integridade da entrada
            bytes(entrada)).hexdigest(),
        "saida_modelo": saida,                   # o que o modelo sugeriu
        "confianca": round(float(confianca), 4),
        "decisao_humana": decisao_humana,        # o que o humano decidiu
        "seguiu_modelo": saida == decisao_humana,
    }
    with open(caminho, "a") as f:
        f.write(json.dumps(evento, ensure_ascii=False) + "\n")
    return evento

# A trilha permite reconstruir, apos o fato, QUEM decidiu O QUE, com QUE
# base e com QUE modelo -- pre-requisito de qualquer responsabilizacao.
print("Funcao registrar_decisao() definida.")
```

```python
# A trilha em acao: uma sessao de triagem em que o humano decide caso a caso
# (o modelo RECOMENDA; o operador SEGUE ou VETA -- controle significativo)
TRILHA = "trilha_auditoria.jsonl"
open(TRILHA, "w").close()                        # comeca uma trilha limpa

# sequencia simulada: (id_contato, classe sugerida, confianca, decisao humana)
# note os casos em que o operador VETA a recomendacao do modelo
sessao = [
    (b"contato-001", "neutro",  0.93, "neutro"),   # segue: alta confianca
    (b"contato-002", "hostil",  0.88, "hostil"),   # segue
    (b"contato-003", "hostil",  0.51, "neutro"),   # VETA: confianca baixa
    (b"contato-004", "neutro",  0.72, "hostil"),   # VETA: contexto humano
    (b"contato-005", "hostil",  0.95, "hostil"),   # segue
    (b"contato-006", "neutro",  0.60, "revisar"),  # VETA: pede 2a analise
]
for entrada, saida, conf, humano in sessao:
    registrar_decisao(TRILHA, "detector-v2.3", entrada, saida, conf, humano)

# reconstrucao apos o fato: quem decidiu o que, com que base?
print("=== TRILHA DE AUDITORIA (reconstruida do registro) ===\n")
eventos = [json.loads(l) for l in open(TRILHA)]
for e in eventos:
    marca = "SEGUIU " if e["seguiu_modelo"] else "VETOU  "
    print(f"[{marca}] modelo sugeriu '{e['saida_modelo']}' "
          f"(conf {e['confianca']:.2f}) -> humano decidiu "
          f"'{e['decisao_humana']}'  | hash {e['entrada_hash'][:12]}...")

taxa_veto = sum(not e["seguiu_modelo"] for e in eventos) / len(eventos)
print(f"\nTaxa de veto humano: {taxa_veto:.0%} -- o operador NAO homologou tudo.")
print("Uma taxa de veto ZERO seria sinal de alerta: vies de automacao (Cap. 4-5).")
```

**Observe:** a trilha registra, para cada decisão, **o que o modelo sugeriu, com que confiança, e o que o humano decidiu** — com o *hash* da entrada garantindo a integridade. Dela se reconstrói, após o fato, *quem decidiu o quê, com que base* — o pré-requisito de qualquer responsabilização.

Note o indicador da última linha: uma **taxa de veto zero** seria um sinal de alerta. Se o operador *sempre* segue o modelo, o controle é nominal, não significativo — é o **viés de automação** em números (o Exercício 10.3 explora o uso da trilha para auditar a responsabilidade).

A **Listagem 10.2** implementa a segunda peça: a **ficha de modelo** (*model card*), que documenta *o que o modelo é, para que serve, onde é válido e o que não se deve esperar dele*:

```python
# Listagem 10.2 - Ficha de modelo (model card): documentacao para governanca.
def gerar_ficha_modelo(caminho, nome, versao, uso_previsto, envelope,
                       metricas, robustez, proveniencia, limitacoes):
    # "Ficha de modelo": documenta o que o modelo E, para que serve,
    # ONDE e valido e o que NAO se deve esperar dele.
    ficha = {
        "nome": nome,
        "versao": versao,
        "uso_previsto": uso_previsto,            # para que foi validado
        "envelope_operacional": envelope,        # condicoes de validade
        "metricas": metricas,                    # desempenho (com a cauda)
        "robustez_adversarial": robustez,        # avaliacao do Cap. 8
        "proveniencia_dados": proveniencia,      # origem e curadoria
        "limitacoes": limitacoes,                # o que NAO usar
    }
    with open(caminho, "w") as f:
        json.dump(ficha, f, ensure_ascii=False, indent=2)
    return ficha

# A ficha acompanha o modelo do desenvolvimento ao descarte. Usar um
# modelo fora de seu envelope documentado e uma decisao -- e, portanto,
# uma responsabilidade.
print("Funcao gerar_ficha_modelo() definida.")
```

```python
# Uma ficha real: o detector de alvos do Capitulo 4, documentado para governanca
ficha = gerar_ficha_modelo(
    caminho="ficha_detector.json",
    nome="Detector de alvos maritimos (VANT)",
    versao="2.3.0",
    uso_previsto="Triagem de embarcacoes em video de VANT para priorizar "
                 "a atencao do analista. NAO autoriza emprego de forca.",
    envelope={
        "sensor": "video optico diurno, VANT a 200-500 m",
        "condicoes": "mar ate estado 3, visibilidade > 5 km",
        "fora_do_envelope": "SAR, noturno, neblina -- degradacao nao validada",
    },
    metricas={
        "mAP@0.5": 0.82, "revocacao_alvo": 0.94,
        "pior_caso_p95_latencia_ms": 28, "taxa_falha": 0.03,
    },
    robustez={
        "PGD_epsilon_0.03": "acuracia cai a 0.61 -- FRAGIL a ataque digital",
        "patch_fisico": "nao avaliado -- exigir red team antes do campo",
    },
    proveniencia="dados sinteticos + reais curados; sem fontes estrangeiras "
                 "nao verificadas (Cap. 6, 8)",
    limitacoes="nao distingue tipos de embarcacao; vies noturno; a DECISAO "
               "sobre qualquer contato permanece humana (controle significativo)",
)
print(json.dumps(ficha, ensure_ascii=False, indent=2))
```

**Observe:** a ficha diz, em um só lugar, **onde o modelo é confiável e onde não é** — inclusive a sua *fragilidade adversarial* (Capítulo 8) e o aviso de que a decisão sobre qualquer contato permanece humana. Empregar o modelo **fora do envelope documentado** deixa de ser um acidente e passa a ser uma **decisão — e, portanto, uma responsabilidade** (o Exercício 10.5 preenche a ficha para outro modelo do livro).

> **📝 Nota** — A governança técnica é a **ponte** entre a engenharia deste livro e a ética deste capítulo. Auditabilidade, rastreabilidade e *accountability* são a **realização concreta** do controle humano significativo e da responsabilização exigida pelo DIH. Sem elas, o compromisso ético não passa de retórica; com elas, torna-se uma **propriedade verificável** do sistema. O que parecia, a princípio, um conjunto de boas práticas de engenharia revela-se, ao final, **uma ética** — a de uma IA de defesa responsável, soberana e centrada no ser humano.

O equilíbrio que o livro procurou sustentar é o de duas exigências que só à primeira vista se opõem. De um lado, **desenvolver capacidade real e soberana** — porque um país que não constrói não governa nem se defende. De outro, **comprometer-se com o uso responsável** — porque capacidade sem contenção é perigo. **As duas não competem: a primeira é o que torna a segunda possível.**

Resta a lição mais simples, e a mais difícil. A IA, neste livro, **não decide**: ela informa, percebe, tria, recomenda. **A decisão — e a responsabilidade que a acompanha — permanece humana.** Isso não é uma fraqueza da tecnologia a ser um dia corrigida; **é a condição de seu uso legítimo**.

## 📋 Resumo do capítulo

- **Não há regime global único** e vinculante para a IA militar, mas um mosaico: a **CCW** da ONU (GGE sobre LAWS, princípios desde 2019), os **princípios da OTAN** e o **AI Act** europeu — que, notavelmente, **exclui os usos militares** de seu escopo.

- O **controle humano significativo** é o eixo do debate: o humano deve ter **compreensão, tempo e capacidade de intervir** para que a decisão seja genuinamente sua — contra o viés de automação. A expressão e a definição são disputadas (*in / on / out of the loop*).

- O **DIH** vincula todo emprego da força: distinção, proporcionalidade, precaução e a cláusula de Martens. **Os limites técnicos tornam-se limites jurídicos**, e a autonomia agrava a **lacuna de responsabilização**.

- O **Brasil** tem atuado por um marco normativo robusto, pela aplicação integral do DIH e pelo controle humano — posição coerente com a END e com a **soberania tecnológica**, que viabiliza a governança em vez de conflitar com ela.

- A **governança técnica** — auditabilidade, rastreabilidade e *accountability* — realiza os princípios éticos e jurídicos, apoiada na reprodutibilidade, na V&V e na documentação (**trilha de auditoria**, **ficha de modelo**). Sem ela, o compromisso é retórica.

- A **síntese da obra**: uma IA de defesa responsável, soberana e centrada no humano. Capacidade soberana e uso responsável **reforçam-se**; e a decisão sobre o uso da força **permanece humana** — não por limitação técnica, mas como **condição de legitimidade**.

## ⚠️ Armadilhas comuns

1. **Supor um único marco regulatório claro.** O cenário é fragmentado e evolui: CCW, OTAN, AI Act (que exclui o militar) e o DIH coexistem. Conheça o mosaico e **ancore-se no DIH**, que se aplica sem exceção.

2. **Tratar o controle humano como formalidade.** Um humano que apenas homologa a sugestão da máquina não exerce controle significativo. Assegure **compreensão, tempo e capacidade real de intervir** — e combata o viés de automação por projeto.

3. **Crer que a conformidade com o DIH é só questão de regras.** Ela depende das **propriedades técnicas** do sistema: robustez, auditabilidade, rastreabilidade. Fragilidade técnica é risco jurídico e moral, não apenas de engenharia.

4. **Opor soberania e responsabilidade.** Não competem: a capacidade soberana é o que permite **auditar, validar e governar** a tecnologia segundo valores próprios. Depender de sistemas estrangeiros opacos sacrifica ambas.

5. **Empregar modelos sem governança técnica.** Sem trilha de auditoria, ficha de modelo e envelope documentado, **não há responsabilização possível**. Usar um modelo fora de seu envelope validado é uma decisão — e uma responsabilidade.

6. **Delegar a decisão sobre o uso da força.** A IA informa, percebe e recomenda; a **decisão e a responsabilidade permanecem humanas**. Não é limitação a superar, mas a **condição do emprego legítimo**.
